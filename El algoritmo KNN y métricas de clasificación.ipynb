```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
os.chdir("C:/Users/odela/Downloads")
```

```python
data= pd.read_csv("Breast-cancer.csv")
data.head()
```

```python
data.drop(["id","Unnamed: 32"],axis=1, inplace= True)
data.head()
```

```python
M= data[data.diagnosis=="M"]
B= data[data.diagnosis=="B"]
```

```python
# Diagrama de dispersión alpha sirve para el grosor de los puntos aquí

plt.scatter(M.radius_mean, M.texture_mean, color= "red", label= "Maligno", alpha=.3)
plt.scatter(B.radius_mean, B.texture_mean, color= "green", label= "Benigno", alpha=.3)
plt.xlabel("radios_mean")
plt.ylabel("texture_mean")
plt.legend()
plt.show()

#En el gráfico se muestra que sí hay un empalme entre maligno y beningo y por eso se requieren usar más variables 
```

```python
# 0 vas a ser igual a Beningno y 1 maligno

data.diagnosis=[1 if each == "M" else 0 for each in data.diagnosis]
data
```

```python
y= data.diagnosis.values
x_data= data.drop(["diagnosis"],axis=1)
```

```python
y
```

```python
x_data
```

```python
# Normalización (en una sola instrucción lo hace para cada columna)
x= (x_data-np.min(x_data))/(np.max(x_data)-np.min(x_data))
x
```

```python
# Base de entrenamiento y prueba

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=.30, random_state= 1)
```

```python
# Modelo KNN de K vecinos más cercanos

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
prediction= knn.predict(x_test)
print("Score:", knn.score(x_test, y_test))
```

```python
y_pred= prediction
y_true= y_test
```

```python
# Creación de matriz de confusión (en la matriz de confunsión los valores de 106 y 57 muestran los valores que se pronosticaron correctos y el 6 y el 2 son los errores
# 106 correctos y 6 mal y 57 correctos y 2 mal
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_true, y_pred)
cm
```

```python
# Visualización de la matrix de confursión
import seaborn as sns
f, ax= plt.subplots(figsize=(5,5))

sns.heatmap(cm, annot= True, linewidths=.5, linecolor= "red", fmt=".0f", ax=ax)
plt.xlabel("Y pronosticada")
plt.ylabel("Y real")
plt.show()
```

```python
# Cálculo de la precisión global

Correctos= cm[0,0] +cm[1,1]
Incorrectos= cm[0,1]+ cm[1,0]
PrecisiónGlobal= Correctos/(Correctos+Incorrectos)
PrecisiónGlobal
```

```python
PrecisiónBenigno= cm[0,0]/(cm[0,0]+cm[1,0])
PrecisiónBenigno
```

```python
PrecisiónMaligna= cm[1,1]/(cm[1,1]+cm[0,1])
PrecisiónMaligna
```

```python

from sklearn.metrics import roc_curve
from sklearn.metrics import auc

y_scores= knn.predict_proba(x_test)

fpr, tpr,threshold= roc_curve(y_test, y_scores[:,1]) #fpr= False positive rate, tpr= True positive rate
roc_auc= auc(fpr,tpr)

plt.title( "Receiver Operating Characteristic")
plt.plot(fpr, tpr, "b",label="AUC= %0.2f" % roc_auc) #Aquí se le esta pidiendo que ponga la etiqueta con dos decimales y sea el porcentaje de roc_auc
plt.legend( loc= "lower right")
plt.plot([0,1], [0,1], "r--")
plt.xlim([0,1])
plt.ylim([0,1]) #limites para y y para x
plt.ylabel("Tasa de positivos verdaderos")
plt.xlabel("Tasa de falsos positivos")
plt.title("Curva ROC del metodo KNN")
plt.show()

# MiENTRAS MÁS GRANDE SEA EL AUC, SIGNIFICA QUE PRONOSTICA MEJOR
# LA PARTECITA DE LA CURVA DE LA LINEA AZUL, ES EL AREA DE LO QUE NOS SE PRONOSTICA BIEN Y EL RESTO ES LO QUE SE PRONOSTICA DE MANERA ACERTADA 
```

```python
neighbors = np.arange(1, 9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

# Loop para diferentes valores de k
for i, k in enumerate(neighbors):
    # Clasificador k-NN para el parametro k
    knn = KNeighborsClassifier(n_neighbors=k)

    # Ajuste del clasificador al dataset de entrenamiento
    knn.fit(x_train, y_train)
    
    # Calculo de precision sobre el dataset de entrenamiento
    train_accuracy[i] = knn.score(x_train, y_train)

    # Calculo de precision sobre el dataset de prueba
    test_accuracy[i] = knn.score(x_test, y_test)
```

```python
# Grafico para encontrar un valor optimo de k
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.title('k-NN: by Number of Neighbors')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
