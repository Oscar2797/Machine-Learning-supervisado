```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
url= "https://archive.ics.uci.edu/ml/machine-learning-databases/communities/communities.data"
crime= pd.read_csv(url,header=None, na_values= ["?"])
crime.head()
```

```python
crime[127].describe() #127 es la variable a predecir
```

```python
plt.figure(figsize=(20,6))
sns.heatmap(crime.isnull(),yticklabels=False, cbar=False,cmap="viridis") #Para saber cuantos valores nulos hay, lo amarillo muestra valores nulos
```

```python
# remover características categóricas

crime.drop([0,1,2,3,4],axis=1, inplace= True)
crime
```

```python
# remover renglones con valores faltantes

crime.dropna(inplace=True)
crime
```

```python
plt.figure(figsize=(20,6))
sns.heatmap(crime.isnull(),yticklabels=False, cbar=False,cmap="viridis") #Para saber cuantos valores nulos hay, lo amarillo muestra valores nulos
```

```python
crime.shape
```

```python
X= crime.drop(127, axis=1)
Y= crime[127]
```

```python
X
```

```python
Y
```

```python
 # División en grupos de entrenamiento (training) y de prueba (testing)

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=.30, random_state=1)
```

```python
# Construcción de un modelo de regresión lineal múltiple

from sklearn.linear_model import LinearRegression
linreg= LinearRegression()
linreg.fit(X_train,Y_train)
```

```python
# Impresión de coeficientes de regresión

print("intercepto:", linreg.intercept_)
print("Coeficiente:", linreg.coef_)
```

```python
# predicciones

Y_predic= linreg.predict(X_test)
Y_predic
```

```python
# Cálculo de indicadores de la bondad de ajuste
from sklearn.metrics import r2_score
from sklearn import metrics
```

```python
# impresión de indicadores de bondad de ajuste

print("Valor de R cuadrada", r2_score(Y_test, Y_predic))
print("Error absoluto medio", metrics.mean_absolute_error(Y_test, Y_predic))
print("Error cuadrático medio", metrics.mean_squared_error(Y_test, Y_predic))
print("Raíz del error cuadrático medio", np.sqrt(metrics.mean_squared_error(Y_test, Y_predic)))
```

# Regresión ridge

```python
#Normalizar datos

from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_train_scaled= scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
# Alpha= 0 (Equivalente a una regresión lineal) con normalización

from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=0)
ridge_reg.fit(X_train_scaled, Y_train)
Y_pred= ridge_reg.predict(X_test_scaled)

# impresión de indicadores de bondad de ajuste

print("Valor de R cuadrada", r2_score(Y_test, Y_pred))
print("Error absoluto medio", metrics.mean_absolute_error(Y_test, Y_pred))
print("Error cuadrático medio", metrics.mean_squared_error(Y_test, Y_pred))
print("Raíz del error cuadrático medio", np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```

```python
# Alpha= .1 con normalización

from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=.1)
ridge_reg.fit(X_train_scaled, Y_train)
Y_pred= ridge_reg.predict(X_test_scaled)

# impresión de indicadores de bondad de ajuste

print("Valor de R cuadrada", r2_score(Y_test, Y_pred))
print("Error absoluto medio", metrics.mean_absolute_error(Y_test, Y_pred))
print("Error cuadrático medio", metrics.mean_squared_error(Y_test, Y_pred))
print("Raíz del error cuadrático medio", np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```

```python
# Examinación de los coeficientes de la regresión ridge con normalización

print("intercepto:", ridge_reg.intercept_)
print("Coeficiente:", ridge_reg.coef_)
```

```python
# Definición de un rango de prueba para Alpha

alpha_range= 10.**np.arange(-2,3)
alpha_range
```

```python
# Definir la mejor Alpha

from sklearn.linear_model import RidgeCV
ridgeregcv= RidgeCV(alphas=alpha_range, scoring= "neg_mean_squared_error")
ridgeregcv.fit(X_train_scaled, Y_train)
ridgeregcv.alpha_
```

```python
# Predicciones con la mejor alpha

Y_predicc= ridgeregcv.predict(X_test_scaled)

# impresión de indicadores de bondad de ajuste

print("Valor de R cuadrada", r2_score(Y_test, Y_predicc))
print("Error absoluto medio", metrics.mean_absolute_error(Y_test, Y_predicc))
print("Error cuadrático medio", metrics.mean_squared_error(Y_test, Y_predicc))
print("Raíz del error cuadrático medio", np.sqrt(metrics.mean_squared_error(Y_test, Y_predicc)))
```

```python
# Examinación de los coeficientes de la regresión ridge con normalización

print("intercepto:", ridgeregcv.intercept_)
print("Coeficiente:", ridgeregcv.coef_)
```

# Regresión de lasso

```python
# Prueba con valor de alpha = .001
```

```python
from sklearn.linear_model import Lasso
lassoreg= Lasso(alpha=.001)
lassoreg.fit(X_train_scaled, Y_train)

# Examinación de los coeficientes de la regresión lasso con normalización (Lasso quita variables insignificantes)

print("intercepto:", lassoreg.intercept_)
print("Coeficiente:", lassoreg.coef_)
```

```python
# Prueba con valor de alpha = .01
from sklearn.linear_model import Lasso
lassoreg= Lasso(alpha=.01)
lassoreg.fit(X_train_scaled, Y_train)

# Examinación de los coeficientes de la regresión lasso con normalización (Lasso quita variables insignificantes)

print("intercepto:", lassoreg.intercept_)
print("Coeficiente:", lassoreg.coef_)
```

```python
# Determinación del grado de bondad de ajuste con alpha .01
y_pred= lassoreg.predict(X_test_scaled)

# impresión de indicadores de bondad de ajuste

print("Valor de R cuadrada", r2_score(Y_test, y_pred))
print("Error absoluto medio", metrics.mean_absolute_error(Y_test, y_pred))
print("Error cuadrático medio", metrics.mean_squared_error(Y_test, y_pred))
print("Raíz del error cuadrático medio", np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))
```

```python
# Selección del valor óptimo de alpha para regresión de lasso

from sklearn.linear_model import LassoCV
lassoregcv= LassoCV(n_alphas=100, random_state=1)
lassoregcv.fit(X_train_scaled, Y_train)
print("Alpha óptimo", lassoregcv.alpha_)
```

```python
# Examinación de los coeficientes de la regresión lasso con normalización (Lasso quita variables insignificantes)

print("intercepto:", lassoregcv.intercept_)
print("Coeficiente:", lassoregcv.coef_)
```

```python
# predicción mediante alpha óptimo lasso

y_pred= lassoregcv.predict(X_test_scaled)

print("Valor de R cuadrada", r2_score(Y_test, y_pred))
print("Error absoluto medio", metrics.mean_absolute_error(Y_test, y_pred))
print("Error cuadrático medio", metrics.mean_squared_error(Y_test, y_pred))
print("Raíz del error cuadrático medio", np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))
```
